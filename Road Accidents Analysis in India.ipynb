```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('C:\\Users\\Pradyot\\Desktop\\road.csv')
df
```

```python
df.info()
# This shows that there are no NaN values
```

```python
months = {"Month":[],"Count":[]}
sum_jan = 0
for i in df['JANUARY']:
    sum_jan = sum_jan+i
months["Month"].append("JANUARY")
months["Count"].append(sum_jan)


sum_feb = 0
for i in df['FEBRUARY']:
    sum_feb = sum_feb+i
months["Month"].append("Feb")
months["Count"].append(sum_feb)


sum_march = 0
for i in df['MARCH']:
    sum_march = sum_march +i
months["Month"].append("March")
months["Count"].append(sum_march)


sum_april = 0
for i in df['APRIL']:
    sum_april = sum_april+i
months["Month"].append("april")
months["Count"].append(sum_april)


sum_may = 0
for i in df['MAY']:
    sum_may = sum_may+i
months["Month"].append("may")
months["Count"].append(sum_may)

sum_june = 0
for i in df['JUNE']:
    sum_june = sum_june+i
months["Month"].append("June")
months["Count"].append(sum_june)


sum_july = 0
for i in df['JULY']:
    sum_july = sum_july+i
months["Month"].append("July")
months["Count"].append(sum_july)


sum_aug = 0
for i in df['AUGUST']:
    sum_aug = sum_aug+i
months["Month"].append("August")
months["Count"].append(sum_aug)


sum_sept = 0
for i in df['SEPTEMBER']:
    sum_sept = sum_sept+i
months["Month"].append("Sept")
months["Count"].append(sum_sept)


sum_oct = 0
for i in df['OCTOBER']:
    sum_oct = sum_oct+i
months["Month"].append("October")
months["Count"].append(sum_oct)


sum_nov = 0
for i in df['NOVEMBER']:
    sum_nov = sum_nov+i
months["Month"].append("November")
months["Count"].append(sum_nov)


sum_dec = 0
for i in df['DECEMBER']:
    sum_dec = sum_dec+i
months["Month"].append("December")
months["Count"].append(sum_dec)
print(months)
```

```python
month_df = pd.DataFrame.from_dict(months)
month_df
```

```python
sns.barplot(x='Count',y='Month',data=month_df)
print("Hence we can see that over the years , the probability of a road accident happening is maximum in the summer period and the winter period")
```

```python
tot = 0
for i in month_df['Count']:
    tot = tot+i
winters = sum_oct+sum_dec+sum_nov+sum_jan+sum_feb
print("Respective are the percentages of road accidents occured in different seasons")
win = (winters/tot)*100
print(win,"% in Winters")
summers = sum_march+sum_april+sum_may
summ = (summers/tot)*100
print(summ,"% in Summers")
rain = 100-summ-win
print(rain,"% in rainy")
```

```python
percent = {"Season":[],"Percent":[]}
percent["Season"].append("Summer")
percent["Percent"].append(26)

percent["Season"].append("Winter")
percent["Percent"].append(41)

percent["Season"].append("Rainy")
percent["Percent"].append(33)

per = pd.DataFrame.from_dict(percent)
print(per)
y=np.array(per['Percent'])
x = np.array(per['Season'])
plt.pie(y,labels=x)
plt.legend()
plt.show()
```
